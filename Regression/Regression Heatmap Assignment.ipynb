```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler

df = pd.read_csv('kc_house_data.csv')
```

```python
df = df.drop(df.columns[[12, 13]], axis=1) #Dropping latitude and longitude
df.head(10)
```

```python
sns.heatmap(df.corr());
```

```python
plt.figure(figsize=(16, 6))
heatmap = sns.heatmap(df.corr(), vmin=-1, vmax=1, annot=True, cmap='crest');
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':12}, pad=12);
```

```python
np.triu(np.ones_like(df.corr()))

#Make is a Triangle Correlation Heatmap
plt.figure(figsize=(16, 6))
# define the mask to set the values in the upper triangle to True
mask = np.triu(np.ones_like(df.corr()))
heatmap = sns.heatmap(df.corr(), mask=mask, vmin=-1, vmax=1, annot=True, cmap='crest')
heatmap.set_title('Triangle Correlation Heatmap', fontdict={'fontsize':18}, pad=16);
```

```python
df.corr()[['price']].sort_values(by='price', ascending=False)
```

```python
plt.figure(figsize=(8, 12))
heatmap = sns.heatmap(df.corr()[['price']].sort_values(by='price', ascending=False), vmin=-1, vmax=1, annot=True, cmap='crest')
heatmap.set_title('Features Correlating with Price', fontdict={'fontsize':18}, pad=16)
```
